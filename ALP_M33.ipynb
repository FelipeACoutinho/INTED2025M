2 - Crie um algoritmo que leia 2 números e imprima na tela o resultado das operações acima.

```python
numero1 = int(input("Digite um número inteiro: "))
numero2 = int(input("Digite um número inteiro: "))

#entrada
soma = numero1 + numero2
subt = numero1 - numero2
mult = numero1 * numero2
div = numero1 / numero2
resto = numero1 % numero2
expo = numero1 ** numero2

#saída
print(f"A soma é: {soma}")
print(f"A subtração é: {subt}")
print(f"A multiplicação é: {mult}")
print(f"A divisão é: {div:.2f}")
print(f"O resto da divisão é: {resto}")
print(f"A exponenciação é: {expo:.2e}")
```

3 - crie um algoritmo, que receba o nome do usuário, e 3 notas, depois calcule a média na variável (media), e imprima na tela a seguinte frase:

O aluno (nome) está com (media) de média!

```python
nome = str(input("Digite o seu nome: "))
nota1 = float(input("Digite a sua primeira nota: "))
nota2 = float(input("Digite a sua segunda nota: "))
nota3 = float(input("Digite a sua terceira nota: "))

#entrada
media = (nota1 + nota2 + nota3) / 3

#saída
print(f"O aluno {nome} está com {media:.2f} de média!")
```

Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas = int(input("Horas trabalhadas: "))
valor = float(input("Digite o valor por hora: "))

if(horas<=40):
  print("Seu salário é de",horas*valor,"reais")
elif(horas>40):
  print("Seu salário é de",(horas-40)*valor*0.5+horas*valor,"reais")
```

## Biblioteca Random

Gera valores aleatórios

```python
import random
```

```python
alt = random.randint(0,100)
tent = 5

while (tent >0):
  numero = int(input("Digite um número entre 0 e 100: "))
  if(numero == alt):
    print("Prabéns! Você acertou!!!")
    break
  elif(numero <alt):
    print("Digite um número maior!")
  else:
    print("Digite um número menor!")
  tent -=1
if (tent ==0):
  print("Suas tentativas acabaram, você perdeu!")
  print("O número era: ", alt)
```

##CALCULADORA


```python
while True:
    print("=========================")
    print("   CALCULADORA INTERATIVA")
    print("=========================")
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")
    print("=========================")

    opcao = input("Escolha uma opção: ")

    # --- OPÇÃO 0: SAIR ---
    if opcao == "0":
        print("Saindo do programa... Até a próxima!")
        break

    # --- OPÇÃO 1: MÚLTIPLOS DE 5 ---
    elif opcao == "1":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        print(f"\nMúltiplos de 5 entre {inicio} e {fim}:")
        for i in range(inicio, fim + 1):
            if i % 5 == 0:
                print(i)
        print()

    # --- OPÇÃO 2: CONTAGEM DE NÚMEROS PARES ---
    elif opcao == "2":
        inicio = int(input("Digite o número inicial: "))
        fim = int(input("Digite o número final: "))

        cont = 0
        for i in range(inicio, fim + 1):
            if i % 2 == 0:
                cont += 1

        print(f"\nQuantidade de números pares entre {inicio} e {fim}: {cont}\n")

    # --- OPÇÃO 3: VERIFICADOR DE NÚMERO PRIMO ---
    elif opcao == "3":
        num = int(input("Digite um número inteiro positivo: "))

        if num <= 1:
            print(f"\nO número {num} NÃO é primo.\n")
        else:
            primo = True
            for i in range(2, num):
                if num % i == 0:
                    primo = False
                    break

            if primo:
                print(f"\nO número {num} é PRIMO.\n")
            else:
                print(f"\nO número {num} NÃO é primo.\n")

    # --- OPÇÃO 4: CONVERSOR DE TEMPERATURA ---
    elif opcao == "4":
        c = float(input("Digite a temperatura em Celsius: "))
        f = c * 1.8 + 32
        print(f"\n{c}°C corresponde a {f:.2f}°F\n")

    # --- OPÇÃO INVÁLIDA ---
    else:
        print("\nOpção inválida! Tente novamente.\n")
```
