2 - Crie um algoritmo que leia 2 números e imprima na tela o resultado das operações acima.

```python
numero1 = int(input("Digite um número inteiro: "))
numero2 = int(input("Digite um número inteiro: "))

#entrada
soma = numero1 + numero2
subt = numero1 - numero2
mult = numero1 * numero2
div = numero1 / numero2
resto = numero1 % numero2
expo = numero1 ** numero2

#saída
print(f"A soma é: {soma}")
print(f"A subtração é: {subt}")
print(f"A multiplicação é: {mult}")
print(f"A divisão é: {div:.2f}")
print(f"O resto da divisão é: {resto}")
print(f"A exponenciação é: {expo:.2e}")
```

3 - crie um algoritmo, que receba o nome do usuário, e 3 notas, depois calcule a média na variável (media), e imprima na tela a seguinte frase:

O aluno (nome) está com (media) de média!

```python
nome = str(input("Digite o seu nome: "))
nota1 = float(input("Digite a sua primeira nota: "))
nota2 = float(input("Digite a sua segunda nota: "))
nota3 = float(input("Digite a sua terceira nota: "))

#entrada
media = (nota1 + nota2 + nota3) / 3

#saída
print(f"O aluno {nome} está com {media:.2f} de média!")
```
